```python
def resize_image(image_path, image_name, folder_path, size=224):
    im = pydicom.dcmread(image_path)
    im = Image.fromarray(im.pixel_array)
    im = im.resize((size, )*2, resample=Image.LANCZOS)
    
    new_path = os.path.join(folder_path, image_name)
    
    im.save(new_path)
```
